```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def model(y, t, k, b):
    y1_0, y2_0 = y
    dy1dt = y2_0
    dy2dt = -k * y1_0 - b * y2_0
    return [dy1dt, dy2dt]
```

```python
y0 = [1.0, 0.0]  
k = 1.0 
b = 0.2 
t = np.linspace(0, 10, 1000) 
```

```python
solutions = odeint(model, y0, t, args=(k, b)) 
y1_values = solutions[:, 0]
y2_values = solutions[:, 1]
```

```python
plt.plot(t, y1_values, label='y1')
plt.plot(t, y2_values, label='y2')
plt.xlabel('Time')
plt.ylabel('y1 / y2')
plt.legend()
plt.title('Coupled Differential Equations')
plt.show()
```
